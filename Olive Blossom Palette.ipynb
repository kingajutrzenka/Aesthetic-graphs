*Olive Blossom Palette*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

*copy it:*

```python
def set_olive_blossom_theme():
    """Presentation style â€“ Olive Blossom Palette"""
    palette = ["#A3A380",  # Olive Petal
               "#D7CE93",  # Golden Clover
               "#EFEBCE",  # Arctic Daisy
               "#D8A48F",  # Rose Blush
               "#BB8588"]  # Peach Blossom

    sns.set_theme(
        style="whitegrid",
        rc={
            "axes.spines.top": False,
            "axes.spines.right": False,
            "axes.titlesize": 14,
            "axes.labelsize": 12,
            "grid.alpha": 0.2,
            "figure.dpi": 120,
            "savefig.dpi": 150,
        }
    )
    sns.set_palette(palette)
    return palette

# set style
palette = set_olive_blossom_theme()
```

```python
np.random.seed(42)
x_train = np.random.normal(0, 1, 300)
y_train = 15 * x_train + np.random.normal(0, 10, 300)

x_test = np.random.normal(0.5, 1, 100)
y_test = 15 * x_test + np.random.normal(0, 10, 100)

coef = np.polyfit(x_train, y_train, 1)
line_x = np.linspace(-3, 4, 100)
line_y = coef[0] * line_x + coef[1]

plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, alpha=0.7, color=palette[0], label="Training set")
plt.scatter(x_test, y_test, alpha=0.7, color=palette[1], label="Test set")
plt.plot(line_x, line_y, color=palette[2], linewidth=2.5, label="Regression line")

plt.xlabel("Feature x")
plt.ylabel("Target variable")
plt.title("Presentation Scatterplot â€“ Olive Blossom Style ðŸŒ¿ðŸŒ¸")
plt.legend()
plt.show()
```
