```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

*Copy it:*

```python
def set_blue_burgundy_protea_theme():
    """Presentation style ‚Äì Blue & Burgundy Protea palette"""
    palette = ["#6271A0", "#791E1C", "#7E7E67", "#F6C6C2", "#72524A"]
    sns.set_theme(
        style="whitegrid",
        rc={
            "axes.spines.top": False,
            "axes.spines.right": False,
            "axes.titlesize": 14,
            "axes.labelsize": 12,
            "grid.alpha": 0.2,
            "figure.dpi": 120,
            "savefig.dpi": 150,
        }
    )
    sns.set_palette(palette)
    return palette
```

```python
palette = set_blue_burgundy_protea_theme()
```

*Sample graphs*

```python

np.random.seed(42)
x_train = np.random.normal(0, 1, 300)
y_train = 15 * x_train + np.random.normal(0, 10, 300)

x_test = np.random.normal(0.5, 1, 100)
y_test = 15 * x_test + np.random.normal(0, 10, 100)

coef = np.polyfit(x_train, y_train, 1)
line_x = np.linspace(-3, 4, 100)
line_y = coef[0] * line_x + coef[1]

plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color=palette[0], label="Training set", marker = 'x')
plt.scatter(x_test, y_test, color=palette[1], label="Test set", marker = 'x')

plt.xlabel("Feature x")
plt.ylabel("Target variable")
plt.title("Presentation Scatterplot ‚Äì Blue-Burgundy-Protea Style")
plt.legend()
plt.show()
```

```python
palette = set_blue_burgundy_protea_theme()

np.random.seed(42)
x_train = np.random.normal(0, 1, 300)
y_train = 15 * x_train + np.random.normal(0, 10, 300)

x_test = np.random.normal(0.5, 1, 100)
y_test = 15 * x_test + np.random.normal(0, 10, 100)

coef = np.polyfit(x_train, y_train, 1)
line_x = np.linspace(-3, 4, 100)
line_y = coef[0] * line_x + coef[1]

plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color=palette[2], label="Training set", marker = 'x')
plt.scatter(x_test, y_test, color=palette[4], label="Test set", marker = 'x')
plt.plot(line_x, line_y, color=palette[3], linewidth=2, label="Regression line")

plt.xlabel("Feature x")
plt.ylabel("Target variable")
plt.title("Presentation Scatterplot ‚Äì Blue-Burgundy-Protea Style")
plt.legend()
plt.show()
```
