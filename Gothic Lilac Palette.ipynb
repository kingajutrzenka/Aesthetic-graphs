```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

*copy it:*

```python
def set_gothic_lilac_theme():
    """Presentation style â€“ Gothic Lilac Palette"""
    palette = ["#080d10",  # Raven Black
               "#571947",  # Gothic Plum
               "#4e3559",  # Faded Iris
               "#9166a6",  # Royal Lilac
               "#c8bbe9"]  # Soft Lilac

    sns.set_theme(
        style="whitegrid",
        rc={
            "axes.spines.top": False,
            "axes.spines.right": False,
            "axes.titlesize": 14,
            "axes.labelsize": 12,
            "grid.alpha": 0.2,
            "figure.dpi": 120,
            "savefig.dpi": 150,
        }
    )
    sns.set_palette(palette)
    return palette

# set style
palette = set_gothic_lilac_theme()
```

```python
# set style
palette = set_gothic_lilac_theme()

# synthetic data
np.random.seed(42)
x_train = np.random.normal(0, 1, 300)
y_train = 15 * x_train + np.random.normal(0, 10, 300)

x_test = np.random.normal(0.5, 1, 100)
y_test = 15 * x_test + np.random.normal(0, 10, 100)

# linear fit
coef = np.polyfit(x_train, y_train, 1)
line_x = np.linspace(-3, 4, 100)
line_y = coef[0] * line_x + coef[1]

# plot
plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, alpha=0.8, color=palette[1], label="Training set")
plt.scatter(x_test, y_test, alpha=0.8, color=palette[2], label="Test set")
plt.plot(line_x, line_y, color=palette[0], linewidth=2.5, label="Regression line")

plt.xlabel("Feature x")
plt.ylabel("Target variable")
plt.title("Presentation Scatterplot â€“ Gothic Lilac Style")
plt.legend()
plt.show()
```
